Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

# 1. Dataset Preparation

```python
np.random.seed(42)

x=np.linspace(0, 10, 40)
y=7*x + 23 + np.random.normal(0, 2, 40)
```

```python
data = pd.DataFrame({'x': x, 'y': y})
```

```python
data
```

```python
n = len(y)
```

# 2. Data Visualization

```python
plt.scatter(x,y,color='blue',label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 3. Cost Function (Mean Squared Error)

```python
def compute_cost(x, y, m, c):
    predictions = m*x + c
    return np.sum((predictions-y)** 2) / n
```

## 4. Gradient Descent

```python
def gradient_descent(X, Y, learning_rate=0.01, epochs=1000):
    m,c = 0, 0

    for _ in range(epochs):
        predictions = m*X + c
        dm = (-2 / n) * np.sum(X * (Y - predictions))
        dc = (-2 / n) * np.sum(Y - predictions)

        m -= learning_rate * dm
        c -= learning_rate * dc

    return m, c

# Train the model
learning_rate = 0.01
epochs = 1000
m, c = gradient_descent(x, y, learning_rate, epochs)
```

## 5. Model Prediction and Visualization

```python
y_pred = m*x + c
plt.scatter(x,y,color='blue',label='Data Points')
plt.plot(x, y_pred,color='red',label='Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 6. Model Evaluation

```python
mae = mean_absolute_error(y,y_pred)
mse = mean_squared_error(y,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y,y_pred)

print(f"Model Evaluation Metrics:\nMAE: {mae:.4f}\nMSE: {mse:.4f}\nRMSE: {rmse:.4f}\nR^2: {r2:.4f}")
```
